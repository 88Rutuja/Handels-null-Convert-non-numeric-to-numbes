 # How to fill nulls 

```python
cr.Loan_Amount_Term.value_counts()
```

```python
cr.Gender = cr.Gender.fillna('Male')

#cr.Gender = cr.Gender.fillna("Male", inplace= True) ----------wrong

#cr.Gender.fillna("Male", inplace= True)     ---------------correct
```

```python
cr.LoanAmount = cr.LoanAmount.fillna(142.51153039832286)    # step 1

# cr.LoanAmount = cr.LoanAmount.fillna(cr.LoanAmount.mean())    # step 2 - always use this method 
```

```python
cr.Dependents = cr.Dependents.fillna(cr.Dependents.mean())
```

```python
cr.Credit_History = cr.Credit_History.fillna(cr.Credit_History.mean())
```

```python
cr.Self_Employed = cr.Self_Employed.fillna("No")
```

```python
cr.Married = cr.Married.fillna("Yes")
```

```python
cr.Loan_Amount_Term = cr.Loan_Amount_Term.fillna(cr.Loan_Amount_Term.mean())
```
